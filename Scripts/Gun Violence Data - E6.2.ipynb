## 2. Creating Subset of the Data

```python
# Create a subset that only includes our numerical values

sub1 = df[['n_participants', 'n_killed', 'n_injured', 'victims','suspects']]
```

## 3. Matrix Heatmap

```python
sub1
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub1.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub1.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub1.shape[1]), sub1.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub1.shape[1]), sub1.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# EXPLANATION - Please see notes below:
# It makes sense that 'suspects' has 3 below 0.0, as a suspect would not be the person we expect to be killed, injured, and definitely not the victim
# The 'n_killed' and 'n_injured' correlation of -0.16 makes sense. You cannot be both.
# 'victims' and 'suspects' having a strong correlation to 'n_participants' makes sense. This is how we track how many people are involved
```

## 6. Create a Categorical Plot

```python
sns.histplot(sub1['n_participants'], bins = 100, kde = True)
```

```python
# Convert 'n_participants' to numeric
sub1['n_participants'] = pd.to_numeric(sub1['n_participants'], errors='coerce')
```

```python
# Create the 'Participants Category' column
sub1['Participants Category'] = pd.cut(sub1['n_participants'], bins=[-np.inf, 2, 6, np.inf], labels=['Low', 'Medium', 'High'])
```

```python
sub1['Participants Category'].value_counts(dropna = False)
```

```python
# Convert NaN to 'High' in the 'Participants Category' column
sub1['Participants Category'] = sub1['Participants Category'].fillna('High')
```

```python
sub1['Participants Category'].value_counts(dropna = False)
```

```python
sub1.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="victims", y="n_killed", hue="Participants Category", data=sub1)
```

```python
# Save the visualization
g.figure.savefig(os.path.join(path, 'participants_categorical_plot.png'))
```

```python
# COMMENTS
# We can see a clear increase in the Participant Category as the victims count increases. There is some overlap near the bottom.
```
